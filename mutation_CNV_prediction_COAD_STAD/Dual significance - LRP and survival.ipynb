```python
import numpy as np
import pandas as pd
import os
import pickle
```

```python
survival_wald_test_pvalue = pd.read_csv('./data/Microbiome_Survival_Analysis_Wald_Test_Pvalue.tsv', 
                                        sep='\t', header=0, index_col=0)
survival_logrank_test_pvalue = pd.read_csv('./data/Microbiome_Survival_Analysis_LogRank_Test_Pvalue.tsv', 
                                        sep='\t', header=0, index_col=0)
survival_lrt_pvalue = pd.read_csv('./data/Microbiome_Survival_Analysis_Likelihood_Test_Pvalue.tsv', 
                                        sep='\t', header=0, index_col=0)
```

```python
display(survival_wald_test_pvalue)
display(survival_logrank_test_pvalue)
display(survival_lrt_pvalue)
```

```python
pvalue_df_dict = {'wald': survival_wald_test_pvalue, 
                  'logrank': survival_logrank_test_pvalue, 
                  'lrt': survival_lrt_pvalue}
cancer_list = ['COAD', 'STAD']

for test_name, pvalue_df in pvalue_df_dict.items():
    for cancer in cancer_list:
        print('===============================================================')
        print(f'{test_name.capitalize()} test on {cancer} microbiome data\n'\
              '# of significant microbial taxa (p <= 0.05):')
        display(np.sum(pvalue_df[[cancer, cancer+'_adj']] <= 0.05))
        print('===============================================================')
```

```python
all_cancer_list = [column for column in survival_wald_test_pvalue.columns if column.find('_')==-1]
```

```python
adj_significance_summary = pd.DataFrame(index=survival_wald_test_pvalue.index)
for cancer in all_cancer_list:
    adj_significance_summary[cancer+'_adj_Wald'] = survival_wald_test_pvalue[[cancer+'_adj']] <= 0.05
    adj_significance_summary[cancer+'_adj_Logrank'] = survival_logrank_test_pvalue[[cancer+'_adj']] <= 0.05
    adj_significance_summary[cancer+'_adj_LRT'] = survival_lrt_pvalue[[cancer+'_adj']] <= 0.05
```

```python
with open('./results/adj_significance_summary.pkl', 'wb') as fp:
    pickle.dump(adj_significance_summary, fp)
adj_significance_summary
```

```python
def survival_sig_genera(disease):
    with open('./results/adj_significance_summary.pkl', 'rb') as fp:
        adj_significance_summary = pickle.load(fp)

    num_of_sig = \
        adj_significance_summary[[disease+'_adj_Wald', disease+'_adj_Logrank', disease+'_adj_LRT']].apply(sum, axis=1)
    
    survival_sig_taxa_name = num_of_sig.index[num_of_sig>0]
    survival_sig_genus_name = [tn.split('__')[-1] for tn in survival_sig_taxa_name]
    if survival_sig_genus_name == []: 
        return []
    else: 
        survival_sig_nona_idx = np.where(np.array(survival_sig_genus_name) != '')[0]
        survival_sig_genus_name_nona = np.array(survival_sig_genus_name)[survival_sig_nona_idx]
        return survival_sig_genus_name_nona
```

```python
survival_sig_genera_dict = {}
for cancer in all_cancer_list:
    survival_sig_genera_dict[cancer] = survival_sig_genera(cancer)
    
with open('./results/survival_sig_genera_dict.pkl', 'wb') as fp:
    pickle.dump(survival_sig_genera_dict, fp)
```

## Compare with LRP results - double significance of LRP and survival analysis

```python
%%time
both_sig_dict = dict()
with open('./results/survival_sig_genera_dict.pkl', 'rb') as fp:
    survival_sig_genera_dict = pickle.load(fp)
for cancer in ['COAD', 'STAD']:
    with open(os.path.join(f'./results/LRP/{cancer}', 'lrp_result_dict.pkl'), 'rb') as f:
        lrp_result_dict = pickle.load(f)
    survival_sig_genus_name_nona = survival_sig_genera_dict[cancer]
    both_sig_dict[cancer] = dict()
    for cov in lrp_result_dict['y_col_names']:
        marker_genera = lrp_result_dict['marker_genera_dict'][cov]
        both_sig_dict[cancer][cov] = list(set(survival_sig_genus_name_nona) & set(marker_genera))
with open('./results/both_sig_dict.pkl', 'wb') as fp:
    pickle.dump(both_sig_dict, fp)
```
